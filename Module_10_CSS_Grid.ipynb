Q3 - Explain the use of grid-auto-row and grid-auto-column using code examples.


In CSS Grid, `grid-auto-rows` and `grid-auto-columns` are used to define the size of rows and columns that are automatically created by the browser when the content exceeds the explicitly defined grid.

1. grid-auto-rows:

This property sets the height of any **automatically generated rows** when there are more items than the specified rows in `grid-template-rows`.

If only one row is defined, and more items are placed that require additional rows, then those rows will take the height defined in `grid-auto-rows`.

```css
.container {
  display: grid;
  grid-template-columns: 100px 100px;
  grid-auto-rows: 80px;
}
```

2. grid-auto-columns:
This property sets the width of any **automatically generated columns** when items overflow the defined columns, especially when `grid-auto-flow: column` is used.

If only one column is defined, and more items are added horizontally, the extra columns will take the width defined in `grid-auto-columns`.

```css
.container {
  display: grid;
  grid-template-rows: 100px;
  grid-auto-flow: column;
  grid-auto-columns: 120px;
}
```
These properties help control the size of dynamically added rows or columns, ensuring consistent layout and spacing even when the number of items exceeds the defined grid structure.

Q5 - Explain the difference between justify-items and justify-self using code examples.


1. justify-items:

`justify-items` is a container-level property
->It is used to align all grid items along the horizontal axis (inline axis)inside their grid area.
-> It applies to every item in the grid unless overridden by `justify-self` on a specific item.
Example:
```css
.container {
  display: grid;
  grid-template-columns: 100px 100px;
  justify-items: center;
}
```
In this case, all grid items will be horizontally centered within their grid cells.

---

2. justify-self:
->`justify-self` is an item-level property.
-> It is used to align a single grid item horizontally within its own cell.
-> It can override the alignment set by `justify-items`.
Example:
```css
.item1 {
  justify-self: end;
}
```
Here, only `.item1` will be aligned to the right end of its grid cell, regardless of the containers `justify-items` value.
Conclusion:
-> Use `justify-items` when you want to apply the same horizontal alignment to all items in the grid.
-> Use `justify-self` when you want to control the horizontal alignment of individual items.
